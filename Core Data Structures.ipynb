```python
• The Goal: An application that lets a user add a task, view all tasks, remove a task, and exit. Store the tasks in a Python List.
```

```python
user_list = []
print("What do you want to do?")
while True:
    print("1 == add task ")
    print("2 == view task ")
    print("3 == remove task ")
    print("4 == quit")
    user_choice = (input("Enter choice: "))
    if user_choice == 1:
        user = input("Enter task: ")
        user_list.append(user)
        print("Task added")
    elif user_choice == 2:
        for task in user_list:
            print (task)
        print(user_list)
    elif user_choice == 3:
        remove_list = input("Remove task: ")
        index = task.index(remove_list)
        user_list.pop(index)
        print(user_list)
        print("Task has been removed")
    elif user_choice == 4:
        print("quit")
    else: 
        print("Nothing to remove!")
        break

        
    
```
